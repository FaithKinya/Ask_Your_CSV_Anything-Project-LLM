```python
def ask_csv(question, df):
    q_lower = question.lower()

    if 'region' in q_lower and 'sales' in q_lower:
        return df.groupby('Region')['Sales'].sum()

    elif 'total sales' in q_lower:
        return df['Sales'].sum()

    elif 'year' in q_lower and 'sales' in q_lower:
        return df.groupby(df['Order Date'].dt.year)['Sales'].sum()

    elif 'category' in q_lower and 'profit' in q_lower:
        return df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

    elif 'product' in q_lower and ('most' in q_lower or 'sold' in q_lower):
        return (
            df.groupby('Product Name')['Quantity']
            .sum()
            .sort_values(ascending=False)
            .head(10)
        )

    else:
        return 'Sorry, I can\'t answer that question yet.'
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
import json

# Loading the model
tokenizer = AutoTokenizer.from_pretrained("TheBloke/vicuna-7B-1.1-HF")
model = AutoModelForCausalLM.from_pretrained(
    "TheBloke/vicuna-7B-1.1-HF",
    device_map="auto",
    torch_dtype=torch.float16
)

# Creating a text generation pipeline
llm = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=100)
```

```python
import json

def interpret_question_llm(question):
    q = question.lower()

    # hardrules (guardrails)
    if "most profitable" in q:
        return {"metric": "profit", "group_by": "category"}

    if "sold the most" in q or "most sold" in q:
        return {"metric": "quantity", "group_by": "product"}

    if "sales per region" in q or "sales by region" in q:
        return {"metric": "sales", "group_by": "region"}

    if "sales per year" in q or "sales by year" in q:
        return {"metric": "sales", "group_by": "year"}

    if "total sales" in q:
        return {"metric": "sales", "group_by": "none"}

    # fallback to llm
    prompt = f"""
Return ONLY a JSON object.
metric: sales | profit | quantity
group_by: region | year | category | product | none

Question: "{question}"
"""

    response = llm(
        prompt,
        max_length=120,
        truncation=True,
        do_sample=False
    )[0]["generated_text"]

    try:
        start = response.find("{")
        end = response.find("}") + 1
        return json.loads(response[start:end])
    except:
        return {"metric": "sales", "group_by": "none"}
```

```python
question = "Which category is most profitable?"
intent = interpret_question_llm(question)
print(intent)
```

```python
def ask_csv_llm(question, df):
    intent = interpret_question_llm(question)

    metric = intent["metric"]
    group_by = intent["group_by"]

    if metric == "sales" and group_by == "region":
        return df.groupby("Region")["Sales"].sum()

    if metric == "sales" and group_by == "year":
        return df.groupby(df["Order Date"].dt.year)["Sales"].sum()

    if metric == "profit" and group_by == "category":
        return df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

    if metric == "quantity" and group_by == "product":
        return (
            df.groupby("Product Name")["Quantity"]
            .sum()
            .sort_values(ascending=False)
            .head(10)
        )

    if metric == "sales" and group_by == "none":
        return df["Sales"].sum()

    return "Intent understood, but analysis not implemented yet."
```

```python
# app.py.code

import streamlit as st
import pandas as pd
import json

# LLM + Guardrails Functions
def interpret_question_llm(question):
    """
    Hybrid function: keyword-based guardrails + optional LLM fallback
    """
    q = question.lower()

    # Hard-coded rules (guardrails)
    if "most profitable" in q:
        return {"metric": "profit", "group_by": "category"}
    if "sold the most" in q or "most sold" in q:
        return {"metric": "quantity", "group_by": "product"}
    if "sales per region" in q or "sales by region" in q:
        return {"metric": "sales", "group_by": "region"}
    if "sales per year" in q or "sales by year" in q:
        return {"metric": "sales", "group_by": "year"}
    if "total sales" in q:
        return {"metric": "sales", "group_by": "none"}

    # Optional LLM fallback (placeholder, can integrate LLaMA)
    return {"metric": "sales", "group_by": "none"}


def ask_csv_llm(question, df):
    """
    Convert a natural language question into Pandas analysis
    """
    intent = interpret_question_llm(question)
    metric = intent["metric"]
    group_by = intent["group_by"]

    if metric == "sales" and group_by == "region":
        return df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
    if metric == "sales" and group_by == "year":
        return df.groupby(df["Order Date"].dt.year)["Sales"].sum()
    if metric == "profit" and group_by == "category":
        return df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
    if metric == "quantity" and group_by == "product":
        return df.groupby("Product Name")["Quantity"].sum().sort_values(ascending=False).head(10)
    if metric == "sales" and group_by == "none":
        return df["Sales"].sum()

    return "Intent understood, but analysis not implemented yet."


# Streamlit App
st.title("LLM-Powered CSV Analytics Assistant")
st.write("Ask natural language questions about the Superstore dataset.")

# Load dataset
@st.cache_data
def load_data():
    df = pd.read_csv("Sample - Superstore.csv", parse_dates=["Order Date", "Ship Date"])
    return df

df = load_data()

# User input
question = st.text_input("Enter your question:")

if question:
    result = ask_csv_llm(question, df)

    # Clean printing
    if isinstance(result, pd.Series):
        result.index.name = None
        for idx, val in result.items():
            if isinstance(val, float):
                st.write(f"{idx}: {round(val,2)}")
            else:
                st.write(f"{idx}: {val}")
    else:
        if isinstance(result, float):
            st.write(round(result,2))
        else:
            st.write(result)
```
